```python
### Use the `glob` library to list the files in a directory.
import glob
```

```python
for directory in glob.glob('C:/Users/it/dataset/multi-directory/*'):
    print(directory)    
```

```python
for file in glob.glob('C:/Users/it/dataset/multi-directory/*/*'):
    print(file)
```

```python
data_frame_list = []
for file in glob.glob('C:/Users/it/dataset/multi-directory/*/*'):
    
    data_frame_list.append(pd.read_csv(file))
```

```python
data_frame_list
```

```python
final_data = pd.concat(data_frame_list) 
```

```python
final_data.head(2)
```
